```python
import numpy as np
import pandas as pd

X_train = np.load('./input/dataset_train.npy')
X_test = np.load('./input/dataset_test.npy')
y_train = np.load('./input/train_labels.npy')
y_test = np.load('./input/test_labels.npy')
```

```python
y_test
```

## Fasttext

```python
length = 200
filter_num = 128
kernel_sizes = [1, 2, 3, 5]
max_features = 20000
embed_size = 300
embedding_matrix1 = np.load('./input/embedding_fasttext.npy')
```

```python
embedding_matrix1[0]
```

```python
from TextCNN import textcnn
model = textcnn(length,filter_num, kernel_sizes, max_features, embed_size, embedding_matrix1)
```

```python
from train import modeltrainer
pth1 = './input/TextCNN/model_fasttext.h5'
model_fasttext = modeltrainer(model, X_train, y_train, 256, 15, pth1)
```

```python
model_fasttext.evaluate(X_test, y_test, batch_size=1024, verbose=1)
```

```python
from keras.models import load_model
model.load_weights('./input/TextCNN/model_fasttext.h5')
y_pred = model.predict(X_test, batch_size=1024, verbose=1)
```

```python
from sklearn.metrics import roc_auc_score
score = roc_auc_score(y_test, y_pred)
score
```
